Let's brute-force this.

We'll treat the states as a `frozenset` of values.

```python
import numpy as np
import itertools
```

```python
all_cards = frozenset(range(1, 11))

def next_states(state):
    for i in all_cards:
        if i not in state:
            yield(state | {i})
```

```python
values = {}
moves = {}
values[all_cards] = sum(all_cards)
print(values)
for n in range(len(all_cards)-1, -1, -1):
    for cards in itertools.combinations(all_cards, n):
        state = frozenset(cards)
        next_values = [0]  # for the joker
        for next_state in next_states(state):
            next_values.append(values[next_state])
        if np.mean(next_values) > sum(state):
            moves[state] = "draw"
            values[state] = np.mean(next_values)
        else:
            moves[state] = "hold"
            values[state] = sum(state)
        
```
